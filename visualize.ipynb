```python

def plot_imgs(full_data):
    
    for i in range(0, len(full_data)):
        plt.subplot(10,10,i)
        
        frame = plt.gca()
        frame.get_xaxis().set_visible(False)
        frame.get_yaxis().set_visible(False)
        heatmap = plt.pcolor(full_data[i], cmap = 'gist_gray')#cmap ='RdBu')#cmap=plt.cm.Blues)



#    plt.show()


f = open("weight_dump.txt", "r")

content = [x.strip('\n') for x in f.readlines()]

# temp_content = [x.strip() for x in f.readlines()]
# temp_content = [x.strip("\n") for x in f.readlines()]
    
# count = 0

# while count < len(temp_content):
#     if temp_content[count] == "":
#         del temp_content[count]
#         count = count - 1
        
#     count += 1
    
    
#print "here", content[1], "hi"
    
full_data = []
for line in content:

    line_split = line.split(" ")
    del line_split[len(line_split)-1]
    
    linear_data = []
    
    for val in line_split:
        linear_data.append(float(val))
    
    data = []
    for i in range(0, 28):
        new_list = []
        data.append(new_list)
        for j in range(0,28):
            data[i].append( linear_data[ i * 28 + j] )
         
    numpy_data = np.array(data)
    full_data.append(numpy_data)
    
print len(full_data)
plot_imgs(np.array(full_data))


```

```python
#Python visualization for the weights
import matplotlib.pyplot as plt
import numpy as np

data = [line.replace('\n','') for line in open('weights_visual.txt')]

for i in range(0,len(data),30):
    data_temp = data[i:i+28]
    for j in range(0,len(data_temp)):
        data_temp[j] = data_temp[j].split(',')
        del data_temp[j][-1]
        data_temp[j] = [float(x) for x in data_temp[j]]
    data_temp = np.array(data_temp)
    #print data_temp
    #fig, ax = plt.subplots()
    plt.subplot(10,10,i/30+1)
    frame = plt.gca()
    frame.get_xaxis().set_visible(False)
    frame.get_yaxis().set_visible(False)
    heatmap = plt.pcolor(data_temp, cmap = 'gist_gray')#cmap ='RdBu')#cmap=plt.cm.Blues)
    
plt.show()
#plt.savefig("HEAT")
```
